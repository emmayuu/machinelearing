# Azure sentiment analysis on Kaggle Comments

```python
import requests
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## API

```python
subscription_key="ae472ec61f5a4509804419fe18dc8776"
assert subscription_key
```

```python
text_analytics_base_url =\
"https://westcentralus.api.cognitive.microsoft.com/text/analytics/v2.0/"
```

```python
documents = {"documents":[]}
path = "/Users/emma/Dropbox (School of Management)/MSBA/all courses/第二学期/452machinelearning/pre/ForumMessages1.txt"
doc1 = open(path, "r")
output = doc1.readlines()
count = 0
for line in output:
    count +=1
    record = {"id": count, "language": "en", "text": line}
    documents["documents"].append(record)

#print it out
documents
```

```python
sentiment_api_url = text_analytics_base_url + "sentiment"
headers   = {"Ocp-Apim-Subscription-Key": subscription_key}
response  = requests.post(sentiment_api_url, headers=headers, json=documents)
sentiments = response.json()
sentiments
```

```python
df = pd.DataFrame(sentiments['documents'])
```

```python
df.head()
df.describe()
```

```python
fig, ax = plt.subplots()
sns.kdeplot(df['score'], shade=True, ax=ax)
fig.set_size_inches(11.7, 8.27)
ax.set_title("Density Plot of Sentiment Analysis for Commends on Kaggles via Azure NLP ")
```
